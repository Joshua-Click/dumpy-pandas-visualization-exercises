```python
x = list(range(1,20))
y1 = [math.sqrt(n) for n in x]
plt.subplot(221)
plt.plot(x, y1)

y2 = [(n**3) for n in x]
plt.subplot(222)
plt.plot(x, y2)


y3 = [2**n for n in x]
plt.subplot(223)
plt.plot(x, y3)

y4 = [(1 / (n+1)) for n in x]
plt.subplot(224)
plt.plot(x, y4)

plt.show()
```
